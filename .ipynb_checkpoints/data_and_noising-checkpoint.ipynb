# Different Noising Distributions

```python
def plot_data(new_data, data, new_data_legend, marker='*', total_data = 10000,
             xlim = None, ylim = None): 
    #noised, eps = noise_gaussian(X, noiselevel)
    plt.scatter(new_data[:total_data, 0], new_data[:total_data, 1], marker=marker, alpha=0.5)
    plt.scatter(data[:total_data, 0], data[:total_data, 1], alpha=0.5)
    plt.legend([new_data_legend, "Original data"])
    if xlim:
        plt.xlim(xlim)
    if ylim:
        plt.ylim(ylim)
    plt.show()

def plot_image(data, cond):
    plt.imshow(data[0], cmap='gray')
    plt.title('{}'.format(cond))
    plt.show()

def denoise_ddpm(noised, eps, noiselevel, baralphas):
    return 1 / torch.sqrt(baralphas[noiselevel]) * (noised - torch.sqrt(1 - baralphas[noiselevel]) * eps)

def denoise_stable(noised, eps, alpha, noiselevel, bargammas):
    return 1 / (bargammas[noiselevel]**(1 / alpha)) * (noised - ((1 - bargammas[noiselevel]) ** (1/alpha)).repeat(1, noised.shape[1]) * eps)
```

# Verification

Do 
$$  \Sigma_t = \sum_{k = 1}^t a_k \left(\dfrac{\bar \gamma_t}{\bar \gamma_k}(1 - \gamma_k)\right)^{2/\alpha} $$
and 
$$ (1 - \bar \gamma_t)^{2/\alpha} a_t \sim \mathcal{S}_{\alpha / 2}(\cdots) $$
have the same distribution?

It seems so.

```python
tmp_diff_steps = 1000
gammas, bargammas = NS.gen_noise_schedule(tmp_diff_steps)

def gen_sigma_a(alpha):
    tmp_skewed_levy_data = DataGenerator.Generator('skewed_levy', alpha = alpha)
    tmp_a_tilde = tmp_skewed_levy_data.generate(size = (tmp_diff_steps))

    tmp_sigma_a = torch.zeros((tmp_diff_steps))
    tmp_a_t = torch.zeros(tmp_diff_steps)
    for i in range(1, tmp_diff_steps):
        tmp_sigma_a[i] = gammas[i]**(2/alpha)*tmp_sigma_a[i-1] \
                + (1 - gammas[i])**(2 / alpha) * tmp_a_tilde[i]
    return tmp_sigma_a

def gen_tmp(alpha):
    tmp_skewed_levy_data = DataGenerator.Generator('skewed_levy', alpha = alpha)
    tmp_a_tilde = tmp_skewed_levy_data.generate(size = (tmp_diff_steps))

    tmp_sigma_a = torch.zeros((tmp_diff_steps))
    tmp_a_t = torch.zeros(tmp_diff_steps)
    for i in range(1, tmp_diff_steps):
        tmp_sigma_a[i] = gammas[i]**(2/alpha)*tmp_sigma_a[i-1] \
                + (1 - gammas[i])**(2 / alpha) * tmp_a_tilde[i]
    for t in range(tmp_diff_steps - 1, 0, -1):
        tmp_a_t[t] = tmp_sigma_a[t] / (1 - bargammas[t])**(2/alpha)
        tmp_a_t_1 = tmp_sigma_a[t-1] / (1 - bargammas[t-1])**(2/alpha)
    return tmp_a_t

def gen_other(t, alpha):
    tmp_skewed_levy_data = DataGenerator.Generator('skewed_levy', alpha = alpha)
    other_a_t_1 = tmp_skewed_levy_data.generate(size = 1)
    other_a_t_prime = tmp_skewed_levy_data.generate(size = 1)
    other_a_t = (1 / (1 - bargammas[t])**(2/alpha))*(gammas[t]**(2 / alpha) \
    *other_a_t_1*(1 - bargammas[t-1])**(2/alpha) \
    + (1 - gammas[t])**(2/alpha)*other_a_t_prime)
    return other_a_t

tmp_a_t = gen_tmp(1.9)
other_a_t = torch.tensor([gen_other(t, 1.9) for t in range(tmp_diff_steps-1, 0, -1)])
```

```python
tmp_sigma_a = torch.stack([gen_sigma_a(1.7) for i in range(200)])
```

```python
other_a_t = torch.tensor([[gen_other(t, 1.7) for t in range(tmp_diff_steps-1, 0, -1)] for j in range(200)])
```
